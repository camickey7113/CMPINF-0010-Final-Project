# Enrollment Statistics

First, let's import pandas and matplot...

```python
import pandas as pd
import matplotlib.pyplot as plt
```

...and bring in the data set.

```python
data = pd.read_csv("school.csv")
data.head(3)
```

As you can see, there is a lot of unneccessary data that makes it hard to see what we need. In order to get rid of this, we can trim the data to only what we need.

```python
data = pd.DataFrame(data, columns=["Neighborhood","Estimate; Total:","Estimate; Enrolled in school:"])
data.head(5)
```

Now, let's make a new column that displays the percentage of the population that attends school by dividing the number enrolled by the total and multiplying by two.

```python
data["ratio"] = round(data["Estimate; Enrolled in school:"] / data["Estimate; Total:"] * 100, 3)
data.head(5)
```

Now that we have the percentages, we can sort them in descending order based on the percentage in order to determine the best neghborhoods in terms of the percentages.

```python
data = data.sort_values("ratio", ascending=False)
data.head(5)
```

```python
data = pd.DataFrame(data, columns=["Neighborhood","ratio"])
data.head(5)
```

Finally, now that we have the data we need, we can display it in a bar graph to visualize the data.

```python
data.plot(x='Neighborhood', y='ratio', kind='bar', width=1)

plt.xlabel('Neighborhood')
plt.ylabel('Ratio')
plt.title('Enrollment by Neighborhood')
plt.show()

data
```
