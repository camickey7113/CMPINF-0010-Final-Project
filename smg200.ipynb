```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv("playing fields.csv")

filtered_df=df[df['field_usage'].str.contains('Diamond and Rectangular')]
print(filtered_df['neighborhood'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("playing fields.csv")

grouped = df.groupby(['neighborhood', 'field_usage'])['field_usage'].count().unstack('field_usage').fillna(0)
plt.figure(figsize=(30,12))
grouped.plot(kind='bar', stacked=True)

plt.xlabel('neighborhood')
plt.ylabel('count')
plt.title('playing fields by neighborhood and type of usage')
plt.show()
```
